```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
psm=("https://www.tn.gov.in/district_view")
```

```python
response=requests.get(psm)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
number=soup.select(".dir_sno")
```

```python
number
```

```python
name=soup.select(".dir_title a")
```

```python
name
```

```python
area=soup.select(".dir_area")
```

```python
area
```

```python
import pandas
```

```python
import pandas as ms
```

```python
districttable=ms.DataFrame(number,columns=["district number"])
```

```python
districttable["district name"]=name
```

```python
districttable["distric area"]=area
```

```python
districttable
```

```python
districttable.to_csv('D:\jupiter notebook\disct.csv')
```
